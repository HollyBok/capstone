```python
# I am going to do a DBSCAN as opposed to a k-means because I am interested in my outliers
# and DBSCAN is less sensitive to that
# However this could be wrong because maybe I DO want a model that is not sensitive 
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN, KMeans
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('datasets/ULTIMATE.csv')
```

```python
df_numeric = df.drop(columns=['player_name', 'team', 'height', 'country',
                             'draft_number_group'])
```

```python
ss = StandardScaler()
df_scaled = ss.fit_transform(df_numeric)
```

```python
dbscan = DBSCAN(eps=3, min_samples=7)
dbscan.fit(df_scaled)
```

```python
set(dbscan.labels_)
```

```python
silhouette_score(df_scaled, dbscan.labels_)
```

```python
df['cluster'] = dbscan.labels_
```

```python
df = pd.read_csv('datasets/ULTIMATE.csv')
df.drop(columns='20_21_salary', inplace=True)
df_numeric = df.drop(columns=['player_name', 'team', 'height', 'country',
                             'draft_number_group', '19_20_salary'])
ss = StandardScaler()
df_scaled = ss.fit_transform(df_numeric)
```

```python
km = KMeans(n_clusters=21)
km.fit(df_scaled)
```

```python
# set(km.labels_)
```

```python
silhouette_score(df_scaled, km.labels_)
```

```python
df['cluster'] = km.labels_
df.to_csv('datasets/ULTIMATECLUSTERED.csv')
df['cluster'].value_counts()
```

```python
clusters = df.groupby(df['cluster'])['19_20_salary'].mean().sort_values(ascending=False)
```
