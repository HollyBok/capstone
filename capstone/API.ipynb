```python
from nba_api.stats.static import players
from nba_api.stats.endpoints import playercareerstats
import pandas as pd
import time

```

## Active Players LIst

```python
nba_players = players.get_players()
```

```python
player_names = []
player_ids = []
player_active = []
```

```python
for each in nba_players:
    player_names.append(each['full_name'])
    player_ids.append(each['id'])
    player_active.append(each['is_active'])
```

```python
players_df = pd.DataFrame({'player_name': player_names,
                          'player_id': player_ids,
                          'active' : player_active})
```

```python
active_players_df = players_df[players_df['active'] == True]
```

```python
active_players_df.head()
```

```python
active_players_df.to_csv('datasets/player_names.csv')
```

## career stats


```python
career_stats = []
errors = []
count = 0
for each in active_players_df['player_id']:
    stats = playercareerstats.PlayerCareerStats(player_id=each)
    stats = stats.get_data_frames()[1]
    try:
        player_career = {'player_id': stats['PLAYER_ID'][0],
                         'league_id' : stats['LEAGUE_ID'][0],
                         'team_id' : stats['Team_ID'][0],
                         'gp' : stats['GP'][0],
                         'gs' : stats['GS'][0],
                         'min' : stats['MIN'][0],
                         'fgm' : stats['FGM'][0],
                         'fga' : stats['FGA'][0],
                         'fg_pct' : stats['FG_PCT'][0],
                         'fg3m' : stats['FG3M'][0],
                         'fg3a' : stats['FG3A'][0],
                         'fg3_pct' : stats['FG3_PCT'][0],
                         'ft_pct' : stats['FT_PCT'][0],
                         'oreb' : stats['OREB'][0],
                         'dreb' : stats['DREB'][0],
                         'reb' : stats['REB'][0],
                         'ast' : stats['AST'][0],
                         'stl' : stats['STL'][0],
                         'blk' : stats['BLK'][0],
                         'tov' : stats['TOV'][0],
                         'pf' : stats['PF'][0],
                         'pts' : stats['PTS'][0], 
                        }
    except:
        errors.append(each)
        print('error on:', each)
    career_stats.append(player_career)
    count += 1
    time.sleep(3)
    if count % 25 == 0:
        print(count)

    
```

```python
df = pd.DataFrame(career_stats)

#columns = keys?
df.to_csv('datasets/career_stats.csv')
```
